## 这道题的关键在于审题和特判
* 前缀和关键点
  * 如果前缀差值从根节点开始怎么办？
    * 将目标差值的字典计入为1
  * 看清楚是不是区间，否则计数在前 还是累加在前
  * 拿一个简单的数基于题目大致给出一个案例
* 


```python
# 朴素写法 自己想的
# Definition for a binary tree node.
# class TreeNode(object):
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution(object):
    def pathSum(self, root, targetSum):
        """
        :type root: Optional[TreeNode]
        :type targetSum: int
        :rtype: int
        """
        def dfs(r,val_sum,t_nums):
            if r==None:
                return 0
            val_sum+=r.val
            tar=val_sum+targetSum

            if val_sum in t_nums:# 如果求的是区间则是累加在前
                counter=t_nums[val_sum]
            else:
                counter=0

            if tar in t_nums:# 如果求的是节点组合则是计数器在前
                t_nums[tar]+=1
            else:
                t_nums[tar]=1
            left=dfs(r.left,val_sum,t_nums)
            right=dfs(r.right,val_sum,t_nums)
            if tar in t_nums:# 前缀和需要回溯
                t_nums[tar]-=1
                
            return left+right+ counter
        
        return dfs(root,0,{targetSum:1})# 如果求的是区间，前缀和最需要注意根节点或者开始区间！从根开始！
    
    
        
            
            
```
