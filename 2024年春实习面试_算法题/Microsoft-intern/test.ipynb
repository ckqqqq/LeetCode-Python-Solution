```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 设置中文显示
plt.rcParams['font.sans-serif'] = ['Arial Unicode MS']


columns_and_index = ['session_daily_cnt', 'Fundamental_dwelltime', 'msn_revenues', 'search_revenue', 'News_show_cnt', 'Search_count', 'scroll_homepage_feed_cnt', 'clicks_tabs_delete_cnt', 'total_mini_apps_open_cnt', 'click_sum', 'pageview_sum']

# 定义数据
matrix_data = [
    [1.00, 0.15, 0.02, 0.02, 0.12, 0.13, 0.03, 0.05, 0.06, 0.11, 0.09],
    [0.15, 1.00, 0.08, 0.06, 0.40, 0.42, 0.19, 0.06, 0.12, 0.25, 0.32],
    [0.02, 0.08, 1.00, 0.01, 0.18, 0.10, 0.09, 0.03, 0.03, 0.09, 0.12],
    [0.02, 0.06, 0.01, 1.00, 0.03, 0.04, 0.01, 0.01, 0.01, 0.03, 0.03],
    [0.12, 0.40, 0.18, 0.03, 1.00, 0.31, 0.27, 0.11, 0.14, 0.29, 0.36],
    [0.13, 0.42, 0.10, 0.04, 0.31, 1.00, 0.04, 0.09, 0.15, 0.31, 0.23],
    [0.03, 0.19, 0.09, 0.01, 0.27, 0.04, 1.00, 0.02, 0.01, 0.08, 0.95],
    [0.05, 0.06, 0.03, 0.01, 0.11, 0.09, 0.02, 1.00, 0.03, 0.26, 0.07],
    [0.06, 0.12, 0.03, 0.01, 0.14, 0.15, 0.01, 0.03, 1.00, 0.12, 0.09],
    [0.11, 0.25, 0.09, 0.03, 0.29, 0.31, 0.08, 0.26, 0.12, 1.00, 0.22],
    [0.09, 0.32, 0.12, 0.03, 0.36, 0.23, 0.95, 0.07, 0.09, 0.22, 1.00]
]

np.random.seed(0)  # 为了结果可重复
cor_data = pd.DataFrame(matrix_data, columns=columns_and_index,index=columns_and_index)


# cor_data=cor_data.corr()//
print(cor_data)
plt.subplots(figsize=(9,9),dpi=1080,facecolor='w')# 设置画布大小，分辨率，和底色
# plt.subplots(figsize=(10, 8), dpi=300, facecolor='w')  # 设置画布大小和分辨率

heatmap=sns.heatmap(cor_data,annot=True, vmax=1, square=True, cmap="Blues", fmt='.2g')#annot为热力图上显示数据；fmt='.2g'为数据保留两位有效数字,square呈现正方形，vmax最大值为1

# 使用seaborn绘制热力图
# heatmap = sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar_kws={"shrink": .5})

# 保存图片
heatmap.get_figure().savefig('correlation_heatmap.png', bbox_inches='tight')

# 显示热力图
plt.show()
# fig.get_figure().savefig('df_corr.png',bbox_inches='tight',transparent=True)#保存图片
#bbox_inches让图片显示完整，transparent=True让图片背景透明
```
