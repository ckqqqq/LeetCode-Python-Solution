4. Median of Two Sorted Arrays

Given two sorted arrays nums1 and nums2 of size m and n respectively, return the median of the two sorted arrays.

The overall run time complexity should be O(log (m+n)).

 

Example 1:

Input: nums1 = [1,3], nums2 = [2]
Output: 2.00000
Explanation: merged array = [1,2,3] and median is 2.
Example 2:

Input: nums1 = [1,2], nums2 = [3,4]
Output: 2.50000
Explanation: merged array = [1,2,3,4] and median is (2 + 3) / 2 = 2.5.
 

Constraints:

nums1.length == m
nums2.length == n
0 <= m <= 1000
0 <= n <= 1000
1 <= m + n <= 2000
-106 <= nums1[i], nums2[i] <= 106

```python
# 官方正解 官方正解不太好理解 不太行
# 官方正解可借鉴的思路
# 中位数，前一段的最大值和后一段的最小值
# 保持len(nums1)<=len(nums2)的设定
# 这道题绝对不是各自去找各自的中位数，而是找合并后的中位数，合并后
class Solution:
    def findMedianSortedArrays(self, nums1: List[int], nums2: List[int]) -> float:
        if len(nums1) > len(nums2):
            return self.findMedianSortedArrays(nums2, nums1)
        # 确保 nums1<=nums2
        infinty = 2**40
        m, n = len(nums1), len(nums2)
        left, right = 0, m
        # median1：前一部分的最大值
        # median2：后一部分的最小值
        median1, median2 = 0, 0

        while left <= right:
            # 前一部分包含 nums1[0 .. i-1] 和 nums2[0 .. j-1]
            # // 后一部分包含 nums1[i .. m-1] 和 nums2[j .. n-1]
            i = (left + right) // 2
            j = (m + n +1) // 2 - i
            
            # nums_im1, nums_i, nums_jm1, nums_j 分别表示 nums1[i-1], nums1[i], nums2[j-1], nums2[j]
            nums_im1 = (-infinty if i == 0 else nums1[i - 1])# 设定为最大和最小
           
            nums_jm1 = (-infinty if j == 0 else nums2[j - 1])
            nums_i = (infinty if i == m else nums1[i])
            nums_j = (infinty if j == n else nums2[j])
            print("mid",i,j,left,right)
            print(nums_im1,nums_i,nums_jm1,nums_j)
            if nums_im1 <= nums_j:
                median1, median2 = max(nums_im1, nums_jm1), min(nums_i, nums_j)
                print("me",nums_im1,nums_i,nums_jm1,nums_j,median1,median2)
                # 前一段的最大值和后一段的最小值
                left = i + 1
            else:
                right = i - 1

        return (median1 + median2) / 2 if (m + n) % 2 == 0 else median1
if __name__=="__main__":
    solution=Solution()
    # print(solution.findMedianSortedArrays([],[1,2,3,4]))
    print(solution.findMedianSortedArrays([1,2,4,5,6],[3,7,8]))# 中位数都是从nums1中取的
    # 4.5
    # print(solution.findMedianSortedArrays([2,2,4,4],[2,2,4,4]))
    # print(solution.findMedianSortedArrays([1,2,3],[4,5,6,7]))
    # print(solution.findMedianSortedArrays([1,2,2],[2,5,6]))
    # print(solution.findMedianSortedArrays([1,3],[2]))
    # print(solution.findMedianSortedArrays([1,2],[4]))
    # print(solution.findMedianSortedArrays([1,2,3,3,100],[1,10,10,11,12]))
# 作者：LeetCode-Solution
# 链接：https://leetcode.cn/problems/median-of-two-sorted-arrays/solution/xun-zhao-liang-ge-you-xu-shu-zu-de-zhong-wei-s-114/
# 来源：力扣（LeetCode）
# 著作权归作者所有。商业转载请联系作者获得授权，非商业转载请注明出处。
```
